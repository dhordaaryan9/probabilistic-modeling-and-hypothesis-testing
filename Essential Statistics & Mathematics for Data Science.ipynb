9. A multiple-choice test consists of 10 questions, each with four possible answers. If a
student randomly guesses on each question, what is the probability of getting at least
8 questions correct?

 Data: 
 
 Number of questions (n) = 10 

 Number of possible answers per question (k) = 4

```python
n = 10
p = 0.25

# We want P(X >= 8). 
# In stats, 'cdf' gives P(X <= k). 
# So, P(X >= 8) is 1 - P(X <= 7)
prob_at_least_8 = 1 - binom.cdf(7,n,p)

prob_8 = binom.pmf(8,n,p)
prob_9 = binom.pmf(9,n,p)
prob_10 = binom.pmf(10,n,p)

print(f"Probablity of getting exactly 8 correct: {prob_8:.6f} ")
print(f"Probablity of getting exactly 9 correct: {prob_9:.6f} ")
print(f"Probablity of getting exactly 10 correct: {prob_10:.6f} ")
print(f"Probablity of getting at least 8 correct: {prob_at_least_8:.6f} same as {1/prob_at_least_8:.1f} ")
```

10. The lifetimes of a certain brand of light bulbs are normally distributed with a mean
of 1000 hours and a standard deviation of 100 hours. What is the probability that a
randomly selected light bulb lasts between 900 and 1100 hours?

 Data:
Mean lifetime (μ) = 1000 hours, Standard deviation (σ) = 100 hours, Lifetime range (lower limit x1,
upper limit x2)

```python
mu = 1000     # mean
sigma = 100   # std
x1 = 900      # lower limit
x2 = 1100     # upper limit

# calculating z score
z1 = (x1 - mu) / sigma
z2 = (x2 - mu) / sigma

print(f"Z-score for 900 hours: {z1}")
print(f"Z-score for 1100 hours: {z2}")

# norm.cdf return the area to the left of the given value
probablity = norm.cdf(x2, mu, sigma) - norm.cdf(x1, mu, sigma)

print(f"Probablity that a bulb lasts between 900 and 1100 hours: {probablity:.4f}")
print(f"Percentage: {probablity * 100:.2f}%")

```

# Insights
    -> Here we can say that 68% of companies bulb are last within this 200-hour window (900 to 1100 hours).

11. A study is conducted to estimate the mean height of a population. A random sample
of 100 individuals is selected, and their heights are measured. Calculate a 95%
confidence interval for the population mean height, given that the sample mean
height is 170 cm and the sample standard deviation is 8 cm.

Data:
Sample size (n) = 100, Sample mean (x̄) = 170 cm, Sample standard deviation (s) = 8 cm, Confidence level = 95%

Explanation: In this problem, we use a sample to estimate the population mean
height. By calculating a confidence interval, we provide a range of plausible
values for the population mean. The 95% confidence level indicates that we are
95% confident that the true population mean height falls within the calculated
interval.

```python
sample_mean = 170
sample_std = 8
n = 100
ci = 0.95

standard_error = sample_std / np.sqrt(n)

# Calculate the Confidence Interval
# We use stats.norm.interval for large samples (n > 30)
ci_lower , ci_upper = stats.norm.interval(ci,loc=sample_mean,scale=standard_error)

print(f"Standard Error: {standard_error}")
print(f"95% Confidence Interval: [{ci_lower:.2f}, {ci_upper:.2f}]")
```

# Insights
    -> Here 95% CI is approximately 168.43 to 171.57 means we 95% sure that average height of people is in this range.

12. A manufacturing company claims that the average weight of its product is 500
grams. To test this claim, a random sample of 25 products is selected, and their
weights are measured. The sample mean weight is found to be 510 grams with a
sample standard deviation of 20 grams. Perform a hypothesis test to determine if
there is evidence to support the company's claim.

 Data: Sample size (n) = 25, Sample mean (x̄) = 510 grams, Sample standard deviation (s) = 20 grams, Population mean (μ) = 500 grams
 
Explanation: In this problem, we are conducting a hypothesis test to assess
whether the sample mean weight provides evidence to support the company's
claim about the population mean weight. The null hypothesis (H0) assumes that
the population mean weight is equal to the claimed value, while the alternative
hypothesis (Ha) suggests otherwise.

```python
# Null Hypothesis: mu = 500 (The company's claim is true).
# Alternative Hypothesis: mu != 500 (The actual weight is different from the claim).

n = 25
mu = 500
x = 510
sample_std = 20

t_test = (x - mu) / (sample_std / np.sqrt(n))

# calculate p-value
degree_of_freedome = n-1
p_value = 2 * (1 - stats.t.cdf(abs(t_test), degree_of_freedome))

print(f"T-statistic: {t_test:.4f}")
print(f"p-value: {p_value:.4f}")
```

# Insights

    T-statistic (2.5000) : This value tells us that our sample mean (510) is 2.5 standard errors away from the claimed mean (500).

    p-value (approx. 0.0196) : Standard practice in data science is to use a significance level ($\alpha$) of 0.05.Since our p-value (0.0196) is less than 0.05, we Reject the Null Hypothesis.

13.A researcher wants to test whether a new teaching method improves student
performance. A random sample of 50 students is divided into two groups: one group
taught using the new method and the other using the traditional method. The average
test scores of the two groups are compared. State the null and alternative hypotheses
for this study.

 Data: Sample size (n) = 50, Test scores of the two groups
 
Explanation: In this problem, we are interested in comparing the means of two
groups (new method vs. traditional method). The null hypothesis (H0) states that
there is no significant difference between the means, while the alternative
hypothesis (Ha) suggests that there is a significant difference

```python
import numpy as np
from scipy import stats

group_new = [85, 88, 90, 92, 78, 85, 89, 91, 84, 87] # New method scores
group_trad = [80, 82, 85, 79, 75, 81, 83, 80, 78, 82] # Traditional method scores

t_stat, p_value_two_tailed = stats.ttest_ind(group_new, group_trad)

p_value_one_tailed = p_value_two_tailed / 2

print(f"T-statistic: {t_stat:.4f}")
print(f"One-tailed P-value: {p_value_one_tailed:.4f}")

alpha = 0.05
if p_value_one_tailed < alpha:
    print("Decision: Reject the Null Hypothesis (The new method works!)")
else:
    print("Decision: Fail to Reject the Null Hypothesis (No significant improvement)")
```

# Insights
    
    The P-value: If the $p$-value is less than $0.05$, we conclude that the difference in scores isn't just a lucky coincidence—the new teaching method actually made a difference.

    

14. A survey is conducted to estimate the proportion of people in a city who support a
particular policy. A random sample of 500 individuals is surveyed, and 320 of them
express support for the policy. Calculate a 90% confidence interval for the
population proportion, given the sample proportion.

 Data: Sample size (n) = 500, Number of successes (x) = 320, Confidence level = 90%

 
Explanation: In this problem, we aim to estimate the population proportion based
on the sample proportion. By constructing a confidence interval, we provide a
range of plausible values for the population proportion. The 90% confidence level
indicates that we are 90% confident that the true population proportion falls
within the calculated interval.

```python
import numpy as np
from scipy import stats

n = 500
x = 320
confidence = 0.90

p_hat = x / n

z_critical = stats.norm.ppf((1 + confidence) / 2)

se = np.sqrt((p_hat * (1 - p_hat)) / n)

me = z_critical * se

ci_lower = p_hat - me
ci_upper = p_hat + me

print(f"Sample Proportion (p-hat): {p_hat:.4f}")
print(f"Critical Value (Z*): {z_critical:.3f}")
print(f"Standard Error: {se:.4f}")
print(f"90% Confidence Interval: [{ci_lower:.4f}, {ci_upper:.4f}]")
```

# insigts

    90% Confidence Interval is approximately [0.6047, 0.6753] means We are 90% confident that the actual percentage of the city's population that supports the policy is between 60.47% and 67.53%.

    Since the entire interval is above 50%, the researcher can confidently state that the majority of the city supports the policy. The "Margin of Error" here is about 3.5%, which is standard for a sample size of 500.


15. A company sells smartphones, and the number of defects per batch follows a
Poisson distribution with a mean of 2 defects. What is the probability of having
exactly 3 defects in a randomly selected batch?

Data: Mean number of defects (λ) = 2, Number of defects (x) = 3

Explanation: The problem involves a discrete distribution (Poisson) because we
are dealing with the count of defects in a batch of smartphones. The Poisson
distribution models the probability of a given number of events occurring within a
fixed interval of time or space.

```python
from scipy.stats import poisson

lam = 2
x = 3

prob_3 = poisson.pmf(x, lam)

print(f"The probability of having exactly 3 defects is: {prob_3:.4f}")
print(f"Percentage: {prob_3 * 100:.2f}%")
```

# insights

    The probability is approximately 0.1804 (or 18.04%).
    Even though the average number of defects is 2, there is still an 18% chance that a specific batch will have 3 defects.

    Business Insight: The company can use this to set its quality thresholds. Since 3 defects happens nearly 1 out of every 5 times, it is not considered a "rare" event. If they want to find truly unusual quality failures, they might look for batches with 6 or 7 defects, where the probability would be much lower.
